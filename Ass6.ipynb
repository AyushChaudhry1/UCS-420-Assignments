<b>Question 1:</b> Ask the user to enter a value (e.g., M) for a mathematical function.
<br>-> Generate x values from -10 to 10 using np.linspace().
<br>-> Compute y values for:
<br>o y= M⋅x^2
<br>o y=M⋅sin(x)
<br>-> Plot both functions in a single figure:
<br>o Use different colors and line styles.
<br>o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M=float(input("Enter the value of M: "))
x=np.linspace(-10,10,200)
y1=M*x**2
y2=M*np.sin(x)

plt.plot(x, y1, 'r-', label=f'y = {M}x²')
plt.plot(x, y2, 'b--', label=f'y = {M}sin(x)')
plt.title(f'Mathematical Functions with M = {M}')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```

<b>Question 2:</b> Create a dataset of five subjects and scores.
<br>-> Convert it into a Pandas DataFrame.
<br>-> Plot the scores using a Seaborn bar plot with:
<br>o Different colors for each bar.
<br>o Annotations on top of each bar.
<br>o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import random

data = {
    'Subjects': ['Math', 'Physics', 'Chemistry', 'Computer Science', 'English']
}

df = pd.DataFrame(data)
df['Scores']=[random.randint(60,100) for i in range(5)]

sns.barplot(x='Subjects', y='Scores', data=df, palette='husl')

for i, score in enumerate(df['Scores']):
    plt.text(i, score, str(score), ha='center', va='bottom')

plt.title('Subject Scores Distribution')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(True, axis='y')
plt.show()
```

<b>Question 3:</b> Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
<br> -> Create a 2x2 subplot layout displaying:
<br>o Line plot showing cumulative sum.
<br>o Scatter plot with random noise.
<br> -> Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt
roll_number = 102317164
np.random.seed(roll_number)
data = np.random.randn(50)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))


cumsum = np.cumsum(data)
ax1.plot(cumsum, 'b-')
ax1.set_title('Cumulative Sum')
ax1.grid(True)
ax1.set_xlabel('Index')
ax1.set_ylabel('Cumulative Sum')

noise = np.random.normal(0, 0.01, 50)
ax2.scatter(range(50), data + noise, alpha=0.6)
ax2.set_title('Scatter Plot with Noise')
ax2.grid(True)
ax2.set_xlabel('Index')
ax2.set_ylabel('Value')

ax3.remove()
ax4.remove()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('company_sales_data.csv')
df['total_profit']
```

```python
sns.lineplot(data=df, x='month_number', y='total_profit')
plt.title('Total Profit Over Months')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
for product in products:
    sns.lineplot(data=df, x='month_number', y=product, label=product)
plt.title('Product Sales Over Months')
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_melted = df.melt(id_vars=['month_number'],
                    value_vars=products,
                    var_name='Product',
                    value_name='Sales')
sns.barplot(data=df_melted, x='Product', y='Sales')
plt.title('Sales Distribution by Product')
plt.ylabel('Total Units Sold')
plt.grid(True)
plt.tight_layout()
plt.show()
```
